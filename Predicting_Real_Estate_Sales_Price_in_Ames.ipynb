## Introduction

Trying to accurately predict the price of a real estate is tricky business. This can often lead to disagreements between home-buyers, home-sellers and real estate agents, which be distressing for all parties involved.

<img src="./images/argument.png" width="500" align="left"/><br>

**The reason it can be difficult to accurately predict the price of real estate is because there are many features in a house that will affect it's sale price**.<br><br>
Many people base their predictions of house sale price on an **Inside View**.<br>
The **Inside View** on a topic involves making predictions based on your understanding of the details of the process<sup>1</sup> which can be easily skewed by personal biases!<br><br>
What we truly need for this issue of housing sale price predictions is an **Outside View**.<br>
The **Outside View** involves ignoring these details and using an estimate based on a class of roughly similar previous cases<sup>1</sup>, and is a clearer, more unbiased approach to the problem at hand.

<img src="./images/inside_outside_view.png" width="500" align="left"/><br>

# <u>1. Problem Statement</u>

There are are many variables that determine how much a home can fetch.<br>

Using the Ames (IA) dataset (train, test), we want to find out which variables matter for home sale prices and produce accurate sale price predictions.<br>

In this study, we will develop **Linear**, **LASSO**, and **RIDGE** regression models and train the models on the train dataset. Success will be determined by how well each model predicts the sale price of the test data set, indicated by a low RMSE value when submitting the predicted sale price to the KAGGLE page.<br>

The best model will be the one with the highest R2 score and lowest RMSE.<br>

This model will help provide the **Outside View**, helping to reduce information asymmetry between potential home-buyers, home-sellers and real estate agents.

### Lot Family

```python
train[train['lot_frontage'].isnull()==True][['lot_frontage','lot_area', 'lot_shape', 'lot_config']]
```

As can be seen from all data about 'Lots', zero Lot Frontage does not mean there are no lots, but simply means that there is **no street connected to the property** at all.

Since the null values in the lot_frontage column indicate that the linear feet of street connected to property is zero, it is tempting to impute these null values with 0.<br>
However...

```python
plt.figure(figsize=(10,7))

sns.regplot(x=train.lot_frontage, y=train.saleprice, ci=None, data=train,
            scatter_kws={'s':12, 'color':'red'},line_kws={'color':'limegreen'});
```

From the figure above, we can see that the higher the lot frontage, the higher the sale price of the house.<br>
For the train data set, there are 330 null values in lot_frontage. That is **more than 15%** of the train entire dataset, which is very significant.<br>
As such, simply impute the nulls lot_frontage with 0, as it will provide very biased data to the model to a very large degree!<br>
We will thus **impute the nulls for lot_frontage with the median** as the "best compromise".

```python
# median lot frontage
print(train.lot_frontage.median())
print(test.lot_frontage.median())
```

```python
# We will impute the missing lot frontage values with the median
# Imputing this now so as not to affect the imputing of numerical nulls with 0 later.
train.lot_frontage.fillna(train.lot_frontage.median(), inplace=True)
test.lot_frontage.fillna(test.lot_frontage.median(), inplace=True)
```

Certain categorical features are on a scale of 'Po' to 'Ex'.<br>
We will convert them to integers. This will allow us to see how correlated they are to the sales price, and also reduce the number of features when we pd.get_dummies later.

We will convert the ordinal features into this scale like so:

- 'Po': 1

## Selecting the Numerical Features to develop our model

### Choose numerical features by correlation

We will use the correlation matrix here to 'weed' out numerical values that do not have a strong correlation to the sale price.<br>
Although LASSO regression would convert coefficients of the the weaker features to 0, this luxury is not present in the RIDGE and simple linear regressions.<br>
We are thus removing the weaker correlated numerical features here to provide the best possible data set for our RIDGE and simple linear regression models.

```python
correlations = train.corr().sort_values(by='saleprice', ascending=False)['saleprice']
correlations
```

We will select all numerical features with greater than 0.3 or lesser than -0.3 correlation, as these are the stronger features.

```python
chosen_numerical = list(correlations[(correlations>=0.3)|(correlations<=-0.3)].index)
chosen_numerical
```

# <u>3. Preprocessing and Modelling</u>

### The Target 'y'

Our target y is obviously the saleprice. From the figure below, we can see that the saleprice is right-skewed.

```python
plt.figure(figsize=(15,5))
sns.histplot(data=final_train, x='saleprice', kde=True, color='hotpink')
plt.axvline(x=final_train['saleprice'].median(), color='darkblue', linestyle = '--', label='Median')
plt.axvline(x=final_train['saleprice'].mean(), color='r', linestyle='--', label='Mean')
plt.legend(loc='upper right')
plt.title('Actual Sale Price Distribution');
```

The tail region on the right potentially acts as outliers which will adversely affect the performance of the model later. This is made worse by the large value of the sale price (hundreds of thousands).<br>
Mean and median also far apart from the peak of the distribution curve.<br>
We will now apply a log transformation over the data to remedy this.

```python
y = np.log1p(final_train.saleprice)
# log1p allows to account for small values
```

```python
plt.figure(figsize=(15,5))
sns.histplot(data=final_train, x=np.log1p(final_train.saleprice), kde=True, color='gold')
plt.axvline(x=y.median(), color='darkblue', linestyle = '--', label='Median')
plt.axvline(x=y.mean(), color='r', linestyle='--', label='Mean')
plt.legend(loc='upper right')
plt.title('np.log Sale Price Distribution');
```

We can see that curve has become a lot smoother. The magnitude of the values are now much smaller (no more than 15) as well.<br>
Mean and median are much closer to the peak of the distribution curve.

The metrics used to evaluate each model's will be:
- Root Mean Squared Error (RMSE) of the actual sales price and predicted sales price.

- R2 Score on the model
    - R2 Score greater than 0.85 is preferred - the closer the R2 score is to 1, the more accurate the model is at predicting the sale price
